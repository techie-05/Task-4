```python
df['diagnosis'] = df['diagnosis'].astype('category',copy=False)
df['diagnosis'].value_counts().plot(kind='bar')
```

```python
y = df['diagnosis']
X = df.drop(columns='diagnosis', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled = scaler.fit(X)
```

```python
X_scaled = scaler.transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.metrics import RocCurveDisplay
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(roc_auc_score(y_test, y_pred))
```

```python
f = RocCurveDisplay.from_estimator(lr, X_test, y_test, name='Logistic Regression')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
```

```python
from sklearn.metrics import accuracy_score
y_pred_prob = lr.predict_proba(X_test)[:, 1]
thresholds = np.arange(0, 1.01, 0.05)
```

```python
def sigmoid(z):
    """
    Sigmoid function.

    Args:
        z: Input value (can be a scalar or an array).

    Returns:
        The sigmoid of z.
    """
    return 1 / (1 + np.exp(-z))
```

```python
best_threshold = 0
best_accuracy = 0

for threshold in thresholds:
    y_pred_thresholded = (y_pred_prob >= threshold).astype(int)
    accuracy = accuracy_score(y_test, y_pred_thresholded)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_threshold = threshold
```

```python
print(f"Best threshold: {best_threshold}")
print(f"Best accuracy: {best_accuracy}")
```

```python
x = np.linspace(-10, 10, 100)
y = sigmoid(x)
plt.plot(x,y)
plt.xlabel("Input (z)")
plt.ylabel("Output (probability)")
plt.title("Sigmoid Function")
plt.grid(True)
plt.show()
```
